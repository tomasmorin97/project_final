```python

def determinar_accion_compra(df):
    acciones = ['ninguna']  # Inicializar con 'ninguna' para el primer día

    for i in range(1, len(df)):
        if df['close'][i] > df['Moving Average'][i] and df['close'][i-1] <= df['Moving Average'][i-1]:
            acciones.append('compra')
        elif df['close'][i] < df['Moving Average'][i] and df['close'][i-1] >= df['Moving Average'][i-1]:
            acciones.append('venta')
        else:
            acciones.append('ninguna')
    
    return acciones

# Asegurarse de que ambos índices sean iguales
df2.reset_index(drop=True, inplace=True)

# Aplicar la función para determinar acciones de compra
df2['Accion'] = determinar_accion_compra(df2)

# Mostrar el DataFrame resultante
df2.tail(60)

```
